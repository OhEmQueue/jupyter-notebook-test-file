```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
raw_data = pd.read_csv('data/MarketingRawData.csv')
```

```python
raw_data.shape
```

```python
sns.set_style("darkgrid")
sns.set(rc = {'figure.figsize':(11,7)})
ax = sns.lineplot(x='Week_ID', y='Revenue', data = raw_data, ci=False)
```

```python
ax = sns.lineplot(x='Week_ID', y='Revenue', data = raw_data, hue = 'Promo', ci=False)
```

```python
ax = sns.lineplot(x='Week_ID', y='Revenue', data = raw_data, hue = 'Promo',style='Promo', ci=False, markers=True)
```

```python
ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data)
```

```python
ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data, ci=False, hue = 'Promo')
```

```python
# Example 2 - Investigating the distribution of Revenue, adding the mean

x = raw_data['Revenue'].values

sns.distplot(x, color = 'blue');

# Calculating the mean
mean = raw_data['Revenue'].mean()

#ploting the mean using matplotlib
plt.axvline(mean, 0,1, color = 'red') # axvline: Axis Vertical Line
```

```python
ax = sns.boxplot(x="Day_Name", y="Revenue", data=raw_data)
```

```python
ax = sns.scatterplot(x="Marketing Spend", y="Revenue", data=raw_data, color = 'green')
```

```python
ax = sns.scatterplot(x="Marketing Spend", y="Revenue", data=raw_data, color = 'green', hue = 'Promo', style = 'Promo')
```

```python
ax = sns.scatterplot(x="Marketing Spend", y="Visitors", data=raw_data, color = 'green', hue = 'Promo', size = 'Revenue',
                    sizes=(20, 200))
```

```python
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=raw_data, height=9, ci=False, line_kws={'color': 'red'})
```

```python
ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=raw_data, height=9, ci=False, hue='Promo', markers=["o", "x", "+"])
```

```python
pc_df = raw_data[['Revenue','Visitors','Marketing Spend']].corr(method ='pearson')
```

```python
ax = sns.heatmap(pc_df)
```
